# ScatterPlot(Numerical-Numerical)  --- Multivariate analysis

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

# Bar Plot(Numerical-Categorical)

```python
titanic.head()
```

```python
sns.barplot(data=titanic,x='pclass',y='age',hue='sex')
```

# Box Plot (Numerical-Categorical)

```python
sns.boxplot(data=titanic,x='sex',y='age',hue='survived')
```

# DistPlot (Numerical-Categorical)

```python
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)
```

# HeatMap(Categorical-Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
(titanic.groupby('embarked')[['survived']].mean()*100)
```

# Clustter Map (Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titanic['sibsp'],titanic['survived']))
```

# PairPlot

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

# LinePlot(Numerical-Numerical)

```python
flights.head()
```

```python
new=flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.lineplot(data=new,x='year',y='passengers')
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
